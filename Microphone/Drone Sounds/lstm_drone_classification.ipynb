```python
import pandas as pd 
import numpy as np 
from IPython.display import Audio
import pickle
import numpy as np
import pandas as pd
import librosa
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from sklearn.model_selection import train_test_split
```

```python
with open('padded_data_labels.pkl', 'rb') as f:
    loaded_data = pickle.load(f)
padded_data = loaded_data['Padded_Data']
labels = loaded_data['Labels']
file_names = loaded_data['File_Name']

print("Pickle dosyasından veri başarıyla yüklendi.")
```

```python
hepsi_df=pd.DataFrame(loaded_data)
```

```python
data = np.stack(hepsi_df["Padded_Data"].to_numpy())
labels = hepsi_df["Labels"].to_numpy()

import librosa
import numpy as np

def extract_mfcc_features(audio_data, sample_rate=22050, n_mfcc=40):
   
    features = []
    for signal in audio_data:
        
        mfcc = librosa.feature.mfcc(y=signal, sr=sample_rate, n_mfcc=n_mfcc)
        mfcc_mean = np.mean(mfcc, axis=1)  
        
    
        feature_vector = np.hstack([mfcc_mean])
        features.append(feature_vector)
    
    return np.array(features)

features = extract_mfcc_features(data, sample_rate=22050)
print(f"Feature shape: {features.shape}")
```

# LSTM

```python
from sklearn.model_selection import KFold
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import matplotlib.pyplot as plt

time_steps = 10
if features.shape[1] % time_steps != 0:
    print("feature dizini tam 10'un katı değil ")
    padded_length = (features.shape[1] // time_steps + 1) * time_steps
    padded_features = np.zeros((features.shape[0], padded_length))
    padded_features[:, :features.shape[1]] = features
    features = padded_features

time_steps = 10  
feature_dim = features.shape[1] // time_steps  
X_lstm = features.reshape(features.shape[0], time_steps, feature_dim)
feature_dim = features.shape[1] // time_steps
X_lstm = features.reshape(features.shape[0], time_steps, feature_dim)


X_train, X_test, y_train, y_test = train_test_split(X_lstm, labels, test_size=0.3, random_state=12,stratify=labels)
k_folds = 5
kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)


class FeatureDataset(Dataset):
    def __init__(self, features, labels):
        self.features = torch.tensor(features, dtype=torch.float32)
        self.labels = torch.tensor(labels, dtype=torch.long)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]


class LSTMClassifier(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, num_classes):
        super(LSTMClassifier, self).__init__()
        self.lstm = nn.LSTM(input_size=input_dim, hidden_size=hidden_dim, num_layers=num_layers, batch_first=True)
        self.fc = nn.Sequential(
            nn.Linear(hidden_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        _, (hn, _) = self.lstm(x)
        out = self.fc(hn[-1])  
        return out


input_dim = feature_dim  
hidden_dim = 64  
num_layers = 2  
num_classes = len(np.unique(labels)) 
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
num_epochs = 20  
batch_size = 16
labels = np.array(labels, dtype=np.int64)

all_fold_metrics = []

for fold, (train_idx, test_idx) in enumerate(kf.split(X_lstm)):
    print(f"Fold {fold + 1}/{k_folds}")
    
    
    X_train, X_test = X_lstm[train_idx], X_lstm[test_idx]
    y_train, y_test = labels[train_idx], labels[test_idx]
    
    train_dataset = FeatureDataset(X_train, y_train)
    test_dataset = FeatureDataset(X_test, y_test)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
    
    model = LSTMClassifier(input_dim, hidden_dim, num_layers, num_classes).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    
    for epoch in range(num_epochs):
        model.train()
        for inputs, targets in train_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        print(f"Fold {fold + 1}, Epoch {epoch + 1}/{num_epochs}, Loss: {loss.item():.4f}")
    
    
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(targets.cpu().numpy())
    
    
    label_names = ["altı", "bebop", "savunma", "membo", "jackal"]
    print(f"Fold {fold + 1} Classification Report:")
    report = classification_report(all_labels, all_preds, target_names=label_names, output_dict=True)
    all_fold_metrics.append(report)
    print(classification_report(all_labels, all_preds, target_names=label_names))
    
    
    conf_matrix = confusion_matrix(all_labels, all_preds, normalize='true')
    disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=label_names)
    disp.plot(cmap=plt.cm.Blues, values_format='.2f')
    plt.title(f"Normalized Confusion Matrix - Fold {fold + 1}")
    plt.xlabel("Tahmin Edilen Etiket")
    plt.ylabel("Gerçek Etiket")
    plt.show()


print("K-Fold Sonuçları:")
for i, report in enumerate(all_fold_metrics):
    print(f"Fold {i + 1}:")
    print(f"Accuracy: {report['accuracy']:.4f}")
```
